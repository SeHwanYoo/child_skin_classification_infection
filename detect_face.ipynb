```python
import os 
from glob import glob
import cv2
import matplotlib.pyplot as plt 
```

```python
# img_name = 'C:/Users/user/Desktop/datasets/Child Skin Disease/Total_Dataset/H2/Angioedema/H2_3255_P2_L0.jpg'
img_name = 'C:/Users/user/Desktop/datasets/Child Skin Disease/Total_Dataset/H2/Angioedema/H2_16736_P6_L0.jpg'
face_img_name = 'D:/datasets/Child Skin Disease/Total_Dataset_infection/Train/infection/H10_3101_P4_L0.jpg'

mouth_cascade_name = 'haarcascade_mcs_mouth.xml'
face_cascade_name = 'haarcascade_frontalface_alt.xml'
```

```python
mouth_cascade = cv2.CascadeClassifier()
face_cascade = cv2.CascadeClassifier()
```

```python
if not mouth_cascade.load(cv2.samples.findFile(mouth_cascade_name)):
    print('--(!)Error loading face cascade')
    exit(0)    
    
if not face_cascade.load(cv2.samples.findFile(face_cascade_name)):
    print('--(!)Error loading face cascade')
    exit(0)    
```

```python
img = cv2.imread(img_name)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

face_img = cv2.imread(face_img_name)
face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB)
plt.imshow(face_img)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.equalizeHist(gray)
plt.imshow(gray)
plt.show()

face_gray = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)
face_gray = cv2.equalizeHist(face_gray)
plt.imshow(face_gray)
plt.show()
```

```python
mouth = mouth_cascade.detectMultiScale(gray)
for (x,y,w,h) in mouth:
    center = (x + w//2, y + h//2)
    #  (,  ,  , , ) 
    img = cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 3)
    
plt.imshow(img)
plt.show()

face = face_cascade.detectMultiScale(face_gray)
for (x,y,w,h) in mouth:
    center = (x + w//2, y + h//2)
    #  (,  ,  , , ) 
    face_img = cv2.rectangle(face_img, (x, y), (x+w, y+h), (0, 255, 0), 3)
    
plt.imshow(face_img)
plt.show()
```
